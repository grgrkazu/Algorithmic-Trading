```python
'''
1. What assets? What dates, etc.
GME = "GameStop Corp" and is a computer game retailer. MVIS = "MicroVision Inc", which is the creator of PicoP® scanning technology.
NFLX = "Netflix", and is a subscription-based streaming service. AMC = "AMC Entertainment Holdings Inc", which is a theater chain store.
All four companies are headquartered in the United States and most are among the largest in the world. 
In this report, I analyzed the stock prices of each company from January 1, 2017 to June 1, 2021.

2. Description of strategy
I have specified 4 algos to define the strategy. First, I set up a simple rule to execute a monthly transaction using "Run Monthly". 
Next, I applied "Select All" to use all the data. Third, I used "WeighEqually" to specify the weight applied to each asset. 
This means that if I have four shares, I will always allocate an equal amount of capital to each share. Finally, I specified that 
the asset weights should be readjusted according to these things I specified. Therefore, I have completed an algorithmic strategy 
to execute monthly transactions in a portfolio with multiple stocks. The strategy is very simple. It sells stocks that have risen in 
price, redistributes profits, and buys stocks that have fallen in price, keeping the holdings of each stock even.

3. Select a graph from the output to include and explain it. What is the main message, etc.
From the "backtest results" graph, we can see that the assets that were 100 have increased to about 2,500. In other words, it can be 
said that this algorithm strategy was effective.　In addition, the interesting result is the fact that the sharp increase in assets has
occurred since January 2021. During this period, the outbreak of a pandemic has a large impact on the economy and health, and it is 
thought that the volatility index is increasing. Therefore, asset rebalancing may yield greater returns as price fluctuations increase.

'''
```

```python
!jt -f anonymous -T -N
import bt
```

```python
data = bt.get('GME, AMC, NFLX, MVIS',
              start='2017-01-01', end='2021--6-01')

Strat = bt.Strategy('TRADE_Monthly',
                    [bt.algos.RunMonthly(),
                     bt.algos.SelectAll(),
                     bt.algos.WeighEqually(),
                     bt.algos.Rebalance()])

test = bt.Backtest(Strat, data)

res = bt.run(test)

res.plot(title="Backtest Result")

res.display()

res.plot_security_weights()

res.plot_histogram()
```
